```python
#Slicing party middle inital and state for cleaning
splitSpace = []
for i in memberID['All Data']:
    splitSpace.append(i.split())
```

```python
#Getting First and Last Name
lastName1 = []
firstName1 =[]
for i in splitSpace:
    lastName1.append(i[0][:-1])
    firstName1.append(i[1])
memberID['First_Name'] = firstName1
memberID['Last_Name'] = lastName1
```

```python
partyState = []
for i in splitSpace:
    if len(i) > 5:
        partyState.append(i[-4:])
    else:
        partyState.append(i[-3:])
party =[]
middleI = []
messyState = []
for i in partyState:
    if len(i) >= 4:
        middleI.append(i[:2])
        messyState.append(i[-2:])
        if i[0] == '(Democratic' or i[0] == '(Republican':
            party.append(i[0])
        else:
            party.append(i[1])
    else:
        middleI.append('NaN')
        party.append(i[0])
        messyState.append(i[-2:])  
```

```python
# Cleaning up party
party1 = []
for i in party:
    party1.append(i[1:])
print(party1[:10])
memberID['Party'] = party1
```

```python
#Cleaning up middle initial adding NaNs when necessary
#This might be weird if there are two middle names
#be sure to inspect data
# print(len(middleI))
# print(middleI[:10])
middleI1 = []
for i in middleI:
    middleI1.append(i)
middleI2 =[]
for i in middleI1:
    if i == 'NaN':
        middleI2.append(i)
    else:
        middleI2.append(i[0])
middleI3 = []
for i in middleI2:
    if i == '(Republican' or i == '(Democratic':
        middleI3.append('NaN')
    else:
        middleI3.append(i)
memberID['Middle'] = middleI3
```

```python
#Cleaning up state
state1 = []
for i in messyState:
    if i[0] == '-':
        state1.append(i[1:])
    else:
        state1.append(i)
state2 = []
for i in range(len(state1)):
    if len(state1[i]) == 2:
        holder = state1[i][0] + ' ' + state1[i][:2][-1][:-1]
        #print(holder)
        state2.append(holder)
#         state2.append(state1[i][0])
#         state2.append(state1[i][:2][-1][:-1])
    else:
        state2.append(state1[i][0][:-1])
memberID['State'] = state2
```
